```python
#webカメラリアルタイム!
# カメラの読込み
# 内蔵カメラがある場合、下記引数の数字を変更する必要あり
cap = cv2.VideoCapture(0)


def create_annotation(img):
    # 検出する。
    detection  = detector.detect_from_imgs(img)
    #print(detection)
    toltal_yen = 0
    # 検出結果を画像に描画する。
    for bbox in detection[0]:
        cv2.rectangle(
            img,
            (int(bbox["x1"]), int(bbox["y1"])),
            (int(bbox["x2"]), int(bbox["y2"])),
            color=(0, 255, 0),
            thickness=2,
        )

        #画像に文字を貼り付ける
        cv2.putText(img,
                text=bbox["class_name"],
                org=(int(bbox["x1"]),int(bbox["y1"])),
                fontFace=cv2.FONT_HERSHEY_SIMPLEX,
                fontScale=1.0,
                color=(0, 255, 0),
                thickness=2,
                lineType=cv2.LINE_4)

        
        toltal_yen += float(bbox["class_name"])
        # confidence = round(bbox["confidence"]*100,1)
        # cv2.putText(img,
        #         text=str(confidence),
        #         org=(int(bbox["x1"])+100,int(bbox["y1"])),
        #         fontFace=cv2.FONT_HERSHEY_SIMPLEX,
        #         fontScale=1.0,
        #         color=(0, 255, 0),
        #         thickness=2,
        #         lineType=cv2.LINE_4)

    
    
    toltal_yen = int(toltal_yen)
    cv2.putText(img,
                text= "total: " + str(toltal_yen) +"yen",
                org=(400,400),
                fontFace=cv2.FONT_HERSHEY_SIMPLEX,
                fontScale=1.0,
                color=(0, 255, 0),
                thickness=2,
                lineType=cv2.LINE_4)
    return img


# 動画終了まで、1フレームずつ読み込んで表示する。
while(cap.isOpened()):
    # 1フレーム毎　読込み
    ret, frame = cap.read()

    # GUIに表示
    frame = create_annotation(frame)
    cv2.imshow("Camera", frame)
    # qキーが押されたら途中終了
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# 終了処理
cap.release()
cv2.destroyAllWindows()
```
